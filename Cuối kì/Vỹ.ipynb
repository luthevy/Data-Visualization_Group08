```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df = pd.read_csv("data.csv")
```

```python
df
```

```python
sns.set_theme(style="ticks")
f, ax = plt.subplots(figsize=(7, 7))
sns.histplot(x = df["City"], color='skyblue',kde = False)
ax.patches[2].set_facecolor('salmon')
```

```python
sns.set_theme(style="ticks")
f, ax = plt.subplots(figsize=(15, 8))
sns.histplot(x = df["Specie"], color='skyblue',kde = False)
ax.patches[4].set_facecolor('salmon')
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
```

```python
plt.figure(figsize = (15,8))
sns.lineplot(x = df['Date'], y = df['max'])
```

```python
new_df = df[df['City']=='Ho Chi Minh City'] 
new_df = df[df['Specie']=='temperature'] 
```

```python
new_df = df.drop(labels=['Date','Country','City','Unnamed: 0','Specie','count','variance','median'], axis=1)
```

```python
df = pd.DataFrame(data = new_df, columns = list(new_df.columns))
plt.figure(figsize = (8,8))
sns.boxplot(x="variable", y="value", data=pd.melt(df))
plt.xlabel("Temperature")
plt.ylabel("Values")
plt.show()
```
